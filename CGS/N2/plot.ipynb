```python
# Energies data
bond_lengths = []
E_DFTs = []

data_file = './scripts/Energies_DFT'
with open(data_file,'r') as file_:
    for line in file_:
        if line.startswith('#'):
            continue
        parts = line.strip().split()
        bond_lengths.append(float(parts[0]))
        
        E_list = [float(x) for x in parts[1:3]]
        E_DFTs.append(E_list)
        print(bond_lengths[-1],E_DFTs[-1])


bond_lengths = []
E_FCIs = []
data_file = './scripts/Energies_FCI'
with open(data_file,'r') as file_:
    for line in file_:
        if line.startswith('#'):
            continue
        parts = line.strip().split()
        bond_lengths.append(float(parts[0]))

        E_list = [float(x) for x in parts[1:3]]
        E_FCIs.append(E_list)
        print(bond_lengths[-1],E_FCIs[-1])
```

```python
# read collected data
#bond_lengths = []
T_asp_linear_gaps = []
T_asp_optimal_gaps = []
gaps = []
data_file = './scripts/data_collected'
with open(data_file,'r') as file_:
    for line in file_:
        if line.startswith('#'):
            continue
        parts = line.strip().split()
        #bond_lengths.append(float(parts[0]))
        T_asp_linear_gaps.append(float(parts[1]))
        T_asp_optimal_gaps.append(float(parts[2]))
        gaps.append(float(parts[3]))
        print(bond_lengths[-1],gaps[-1], T_asp_linear_gaps[-1], T_asp_optimal_gaps[-1])
```

```python
# read LK data
L_for_gaps = []
K_for_gaps = []
gaps_for_LK = []
data_file = './scripts/LK_collected'
with open(data_file,'r') as file_:
    for line in file_:
        if line.startswith('#'):
            continue
        parts = line.strip().split()
        gaps_for_LK.append(float(parts[1]))
        L_for_gaps.append(float(parts[2]))
        K_for_gaps.append(float(parts[3]))
        print(gaps_for_LK[-1],L_for_gaps[-1],K_for_gaps[-1])
# sort for better plot

gaps_for_LK = np.array(gaps_for_LK)
L_for_gaps = np.array(L_for_gaps)
K_for_gaps = np.array(K_for_gaps)

sort_indices = np.argsort(gaps_for_LK)
gaps_for_LK = gaps_for_LK[sort_indices]
L_for_gaps = L_for_gaps[sort_indices]
K_for_gaps = K_for_gaps[sort_indices]
```

```python
costs = [4/gap for gap in gaps]
T_asp_optimal_plus_costs = [T_asp_optimal_gaps[i] + costs[i] for i in range(len(gaps))]
```

```python
# read data
bond_length =3.5
asp_schedule_dir = './data/' +str(bond_length) + '/on_the_fly/'
asp_optimal_dir = './data/' +str(bond_length) + '/asp/optimal'
asp_linear_dir = './data/' +str(bond_length) + '/asp/linear'
t_list = []
s_list = []
with open(asp_schedule_dir + '/optimal_schedule','r') as file_:
    for line in file_:
        if not line.strip():
            continue
        if line.startswith('#'):
            continue
        ls = line.split()
        t_list.append(float(ls[0]))
        s_list.append(float(ls[1]))
        #print(ls[0], ls[1])

#print('--')
    
tf_grid_optimal = []
fidelity_optimal = []
tf_grid_linear = []
fidelity_linear = []

with open(asp_optimal_dir + '/fidelity','r') as file_:
    for line in file_:
        if not line.strip():
            continue
        if line.startswith('#'):
            continue
        ls = line.split()
        tf_grid_optimal.append(float(ls[0]))
        fidelity_optimal.append(float(ls[1]))
        #print(ls[0], ls[1])

#print('--')
with open(asp_linear_dir + '/fidelity','r') as file_:
    for line in file_:
        if not line.strip():
            continue
        if line.startswith('#'):
            continue
        ls = line.split()
        tf_grid_linear.append(float(ls[0]))
        fidelity_linear.append(float(ls[1]))
        #print(ls[0], ls[1])
```

```python
# find 75% fidelity time
from scipy.interpolate import interp1d
fidelity_linear_fct = interp1d(tf_grid_linear, fidelity_linear)
fidelity_optimal_fct = interp1d(tf_grid_optimal, fidelity_optimal)

from scipy.optimize import brentq

t_75_linear = brentq(lambda x: fidelity_linear_fct(x) - 0.75, tf_grid_linear[0], tf_grid_linear[-1])
print(t_75_linear)

t_75_optimal = brentq(lambda x: fidelity_optimal_fct(x) - 0.75, tf_grid_optimal[0], tf_grid_optimal[-1])
print(t_75_optimal)

print(t_75_linear/t_75_optimal)
```

```python
from scipy.optimize import curve_fit
def func(x, r, c):
    return c/x**r
f_1 = lambda x, c : func(x, 1, c)
f_2 = lambda x, c : func(x, 2, c)


# only fiter gap with <0.1
idx = [i for i, gap in enumerate(gaps) if gap < 0.1]
gaps_filtered = [gaps[i] for i in idx]
print(gaps_filtered)
T_asp_optimal_filtered = [T_asp_optimal_gaps[i] for i in idx]
T_asp_optimal_plus_costs_filtered = [T_asp_optimal_plus_costs[i] for i in idx]
T_asp_linear_filtered = [T_asp_linear_gaps[i] for i in idx]

optimal_fit, _ = curve_fit(f_1, gaps_filtered, T_asp_optimal_filtered, p0=np.pi/2)
cost_fit, _ = curve_fit(f_1, gaps_filtered, T_asp_optimal_plus_costs_filtered, p0=np.pi/2)
linear_fit, _ = curve_fit(f_2, gaps_filtered, T_asp_linear_filtered, p0=np.pi/2)

print(*optimal_fit, *linear_fit, *cost_fit, sep=', ')
```
