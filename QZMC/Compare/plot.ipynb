```python
# Number of Trotter steps
def NumberOfTrotterSteps(alpha, n_site, which):
    beta = 3
    Uc   = 5.0
    t_inters = np.linspace(0.0,1.0,n_site+1)
    n_dimer = n_site//2
    if which=='Chain':
        n_inter = n_dimer
    elif which=='Ladder':
        n_inter = n_dimer * 2
    trotter_factor = (n_dimer + t_inters[alpha] * n_inter)
    return max(20,int(trotter_factor/8 * 75 * beta**2/3.0**2 * Uc/5.0))
print('# of Trotter Steps for each alpha')
for n_site in [6, 8, 10]:
    n_trotters = np.zeros((n_site+1),dtype=float)
    for alpha_ in range(1,n_site+1):
        n_trotters[alpha_] = NumberOfTrotterSteps(alpha_,n_site,'Chain')
        #print(n_trotters[alpha_])
    print(str(n_site)+' Chain =' + str(2*np.sum(n_trotters[1:])))

    for alpha_ in range(1,n_site+1):
        n_trotters[alpha_] = NumberOfTrotterSteps(alpha_,n_site,'Ladder')
        #print(n_trotters[alpha_])
    print(str(n_site)+' Ladderr =' + str(2*np.sum(n_trotters[1:])))
```
